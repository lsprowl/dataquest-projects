```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from pandas.tools.plotting import scatter_matrix 
```

```python
recent_grads = pd.read_csv('recent-grads.csv')
print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
print(recent_grads.describe())
```

```python
raw_data_count = len(recent_grads.index)
recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads.index)
print(raw_data_count,cleaned_data_count)
```

```python
recent_grads.plot(x='Sample_size',y='Total',kind='scatter',xlim=(0,4500),ylim=(0,450000))
print(recent_grads[(recent_grads['Total']>250000) & (recent_grads['Sample_size']<3000)])
```

```python
recent_grads.plot(x='Women',y='Men',kind='scatter',ylim=(0,200000),xlim=(0,350000))
```

```python
recent_grads['Median'].hist(bins=24,range=(0,120000))
```

```python
recent_grads['ShareWomen'].hist(bins=10,range=(0,1))
```

```python
recent_grads['Unemployment_rate'].hist(bins=20,range=(0,0.2))
```

```python
scatter_matrix(recent_grads[['Total','Sample_size','Men','Women','Employed','Median','Low_wage_jobs','Full_time','Part_time']],figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Total','Employed','Full_time']])
```

```python
recent_grads[:10].plot.bar(x='Major',y='ShareWomen',legend=False,title="Share of women in the 10 highest paying majors")
```

```python
recent_grads[-10:].plot.bar(x='Major',y='ShareWomen',legend=False,title="Share of women in the 10 lowest paying majors")
```

```python
recent_grads.sort_values('ShareWomen')[:10].plot.bar(x='Major',y='ShareWomen',legend=False,title='10 majors with the lowest share of women')
```

```python
recent_grads.sort_values('ShareWomen')[-10:].plot.bar(x='Major',y='ShareWomen',legend=False,title='10 majors with the highest share of women')
```

# Conclusions

The sample size, for the most part, linearly correlates with the total number of people in a major, which is great for having a representative sample. However, two majors, biology and psychology, are under-represented in the sample size.

There is a strong correlationg between the total number of people with a major and the number of people employed meaning that people in more popular majors are more likely to be employed, which is probably why the majors are so popular in the first place. There is also a strong correlation between the number of full-time employees and the number of people employed meaning that most people who are employed are employed full time. The median salary of full-time employees does not correlate well with any other category meaning that none of these categories is indicative of an employee's salary.

The ten highest paying majors are composed mainly of men, while the ten lowest paying majors are composed mainly of women. Most majors with a low share of women are engineering majors. Most majors with a high share of women are education majors or medical assistant related majors.
