```python
# function from https://deparkes.co.uk/2016/11/04/sort-pandas-boxplot/
def boxplot_sorted(df, by, column, rot=0):
    # use dict comprehension to create new dataframe from the iterable groupby object
    # each group name becomes a column in the new dataframe
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    # find and sort the median values in this new dataframe
    meds = df2.median().sort_values()
    # use the columns in the dataframe, ordered sorted by median value
    # return axes so changes can be made outside the function
    return df2[meds.index].boxplot(rot=rot, return_type="axes")

#recent_grads.boxplot(by='Major_category',column='Median',rot=90)
axes = boxplot_sorted(recent_grads,by='Major_category',column='Median',rot=90)
axes.set_title('Boxplot of median salary by major category')
```

```python
# function adapted from https://deparkes.co.uk/2016/11/04/sort-pandas-boxplot/
def boxplot_reverse_sorted(df, by, column, rot=0):
    # use dict comprehension to create new dataframe from the iterable groupby object
    # each group name becomes a column in the new dataframe
    df2 = pd.DataFrame({col:vals[column] for col, vals in df.groupby(by)})
    # find and sort the median values in this new dataframe
    meds = df2.median().sort_values(ascending=False)
    # use the columns in the dataframe, ordered sorted by median value
    # return axes so changes can be made outside the function
    return df2[meds.index].boxplot(rot=rot, return_type="axes")

#recent_grads.boxplot(by='Major_category',column='ShareWomen',rot=90)
axes = boxplot_reverse_sorted(recent_grads,by='Major_category',column='ShareWomen',rot=90)
axes.set_title('Boxplot of share of women by major category')
```

# Conclusions

The sample size, for the most part, linearly correlates with the total number of people in a major, which is beneficial for having a representative sample. However, two majors, biology and psychology, are under-represented in the sample size.

There is a strong correlationg between the total number of people with a major and the number of people employed. This means that people in more popular majors are more likely to be employed, which is probably why the majors are so popular in the first place. There is also a strong correlation between the number of full-time employees and the number of people employed meaning that most people who are employed are employed full time. The median salary of full-time employees does not correlate well with any other category meaning that none of these categories is indicative of an employee's salary.

The ten highest paying majors are composed mainly of men, while the ten lowest paying majors are composed mainly of women. Most majors with a low share of women are engineering majors. Most majors with a high share of women are education majors or health related majors.
